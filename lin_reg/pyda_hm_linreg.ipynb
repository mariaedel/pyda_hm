```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('water.csv')
df.head()
```

```python
df.plot(kind='scatter', x='hardness', y='mortality')
plt.show()
```

```python
df[['hardness', 'mortality']].corr() #коэффициент корреляции Пирсона
```

```python
df[['hardness', 'mortality']].corr(method='spearman') #коэффициент корреляции Спирмена
```

```python
x = df[['hardness']]
y = df['mortality']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=40)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
model.score(x_test, y_test) #коэффициент детерминации
```

```python
plt.scatter(y_test - y_pred, y_pred)
plt.title('График остатков')
plt.show()
```

```python
df_south = df[ df.location=='South' ]
df_north = df[ df.location=='North' ]
```

### SOUTH

```python
df_south.plot(kind='scatter', x='hardness', y='mortality')
plt.show()
```

```python
df_south[['hardness', 'mortality']].corr()
```

```python
df_south[['hardness', 'mortality']].corr(method='spearman')
```

```python
x_south = df_south[['hardness']]
y_south = df_south['mortality']

x_train_south, x_test_south, y_train_south, y_test_south = train_test_split(x_south, y_south, test_size=0.30, random_state=40)

model = LinearRegression()
model.fit(x_train_south, y_train_south)
y_pred_south = model.predict(x_test_south)
model.score(x_test_south, y_test_south)
```
